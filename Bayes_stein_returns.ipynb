```python
def bayes_stein_expected_returns(window_returns):
    """
    Calculate Bayes-Stein shrinkage expected returns for a given window.
    
    Parameters:
    -----------
    window_returns : pandas DataFrame or numpy array
        Shape: (42 days, 46 stocks)
        Each row is a day, each column is a stock's daily return
    
    Returns:
    --------
    shrunk_returns : numpy array
        Shape: (46,) - Expected return for each stock for the next day
    """
    
    # Convert to numpy array if DataFrame
    if isinstance(window_returns, pd.DataFrame):
        returns = window_returns.values
    else:
        returns = window_returns
    
    T = returns.shape[0]  # Number of days (42)
    N = returns.shape[1]  # Number of stocks (46)
    
    # STEP 1: Calculate sample mean for each stock (μ̂)
    mu_hat = np.mean(returns, axis=0)  # Shape: (46,)
    
    # STEP 2: Calculate grand mean (μ̄)
    mu_bar = np.mean(mu_hat)  # Single number
    
    # STEP 3: Calculate covariance matrix (Σ)
    cov_matrix = np.cov(returns, rowvar=False)  # Shape: (46, 46)
    
    # STEP 4: Calculate Spread Measure (D)
    # Step 4a: Calculate deviations (μ̂ - μ̄)
    deviations = mu_hat - mu_bar  # Shape: (46,)
    
    # Step 4b: Invert covariance matrix (Σ⁻¹)
    cov_inv = np.linalg.inv(cov_matrix)  # Shape: (46, 46)
    
    # Step 4c: Calculate D = (μ̂ - μ̄)' × Σ⁻¹ × (μ̂ - μ̄)
    D = deviations.T @ cov_inv @ deviations  # Single number
    
    # STEP 5: Calculate shrinkage weight (w)
    w = (N + 2) / ((N + 2) + T * D)  # Single number
    
    # STEP 6: Apply shrinkage to each stock
    # Formula: μᵢ,shrunk = w × μ̄ + (1-w) × μ̂ᵢ
    shrunk_returns = w * mu_bar + (1 - w) * mu_hat  # Shape: (46,)
    
    return shrunk_returns
```

```python
df = df.sort_values('Date').reset_index(drop=True)

# Parameters
window_size = 42  # Number of trading days in each window
start_date = '2011-11-01'

# Filter data from start_date onwards
df_filtered = df[df['Date'] >= start_date].copy()

# Get return columns (exclude Date)
return_columns = [col for col in df_filtered.columns if col != 'Date']

# Storage for predictions
predictions = []

# Calculate how many windows we can create
total_rows = len(df_filtered)
num_windows = total_rows - window_size

print(f"Total trading days available: {total_rows}")
print(f"Number of windows to create: {num_windows}")
print(f"Window size: {window_size} days\n")

# Create rolling windows
for i in range(num_windows):
    # Define window indices
    window_start_idx = i
    window_end_idx = i + window_size
    
    # Extract window data
    window_data = df_filtered.iloc[window_start_idx:window_end_idx]
    window_returns = window_data[return_columns]
    
    # Get the prediction date (next day after window)
    prediction_date = df_filtered.iloc[window_end_idx]['Date']
    
    # Calculate Bayes-Stein expected returns
    expected_returns = bayes_stein_expected_returns(window_returns)
    
    # Store results
    predictions.append({
        'Window': i + 1,
        'Window_Start': window_data['Date'].iloc[0],
        'Window_End': window_data['Date'].iloc[-1],
        'Prediction_Date': prediction_date,
        'Expected_Returns': expected_returns
    })
    
    # Print progress every 50 windows
    if (i + 1) % 50 == 0:
        print(f"Processed {i + 1} windows...")

# Convert to DataFrame
predictions_df = pd.DataFrame(predictions)

print(f"\nCompleted! Total predictions: {len(predictions_df)}")
print("\nFirst few predictions:")
print(predictions_df[['Window', 'Window_Start', 'Window_End', 'Prediction_Date']].head(10))
```

```python
# Create expected_returns DataFrame
expected_returns = pd.DataFrame()

# Add the Prediction_Date as the Date column
expected_returns['Date'] = predictions_df['Prediction_Date']

# Add each stock's expected returns as separate columns
for i, stock in enumerate(return_columns):
    expected_returns[stock] = predictions_df['Expected_Returns'].apply(lambda x: x[i])

# Reset index
expected_returns = expected_returns.reset_index(drop=True)
print("Expected Returns DataFrame:")
expected_returns.head()
```
